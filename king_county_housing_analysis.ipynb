## Instrucciones

1. Importar las librerías necesarias y suprimir advertencias.


```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import Ridge
from IPython.display import display

%matplotlib inline
```

## Módulo 1: Importación de Datos

### Descargar y cargar el dataset

1. Descargar el dataset desde el enlace proporcionado.
2. Cargar el archivo CSV en un DataFrame de Pandas.
3. Mostrar las primeras filas del DataFrame.
4. Obtener un resumen estadístico del DataFrame.


```python
import requests

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
response = requests.get(url)

# Guardar el archivo
with open('housing.csv', 'wb') as file:
    file.write(response.content)

# Cargar el CSV
df = pd.read_csv('housing.csv')

# Mostrar las primeras filas del dataframe
display(df.head())

# Obtener un resumen estadístico del dataframe
display(df.describe())
```

## Módulo 2: Manipulación de Datos

### Eliminar columnas innecesarias

1. Eliminar las columnas "id" y "Unnamed: 0" del DataFrame.
2. Obtener un resumen estadístico después de eliminar las columnas.


```python
df.drop(["id", "Unnamed: 0"], axis=1, inplace=True)

# Obtener un resumen estadístico después de eliminar las columnas
display(df.describe())
```

### Manejo de valores faltantes

1. Verificar y reemplazar los valores faltantes en las columnas 'bedrooms' y 'bathrooms' con la media de cada columna.
2. Imprimir el conteo de valores NaN antes y después de reemplazarlos.


```python
print("number of NaN values for the column bedrooms:", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms:", df['bathrooms'].isnull().sum())

mean_bedrooms = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan, mean_bedrooms, inplace=True)

mean_bathrooms = df['bathrooms'].mean()
df['bathrooms'].replace(np.nan, mean_bathrooms, inplace=True)

print("number of NaN values for the column bedrooms:", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms:", df['bathrooms'].isnull().sum())
```

## Módulo 4: Desarrollo de Modelos

### Ajustar un modelo de regresión lineal con `long`

1. Ajustar un modelo de regresión lineal usando la característica `long` y calcular el R².


```python
X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X, Y)
print("R^2 de la regresión con 'long':", lm.score(X, Y))
```

### Ajustar un modelo de regresión lineal con `sqft_living`

1. Ajustar un modelo de regresión lineal usando `sqft_living` y calcular el R².

```python
X = df[['sqft_living']]
lm.fit(X, Y)
print("R^2 de la regresión con 'sqft_living':", lm.score(X, Y))
```

### Ajustar un modelo de regresión lineal con múltiples características

1. Ajustar un modelo de regresión lineal usando una lista de características seleccionadas y calcular el R².

```python
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]
X = df[features]
lm.fit(X, Y)
print("R^2 de la regresión con múltiples características:", lm.score(X, Y))
```

### Crear y ajustar un pipeline

1. Crear una lista de tuplas para el pipeline con estimadores de escalado, transformación polinómica y regresión lineal.
2. Ajustar el pipeline con las características seleccionadas y calcular el R².

```python
pipeline_steps = [
    ('scale', StandardScaler()),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
]

pipeline = Pipeline(steps=pipeline_steps)
pipeline.fit(X, Y)
print("R^2 del pipeline con características polinómicas:", pipeline.score(X, Y))
```

## Módulo 5: Evaluación y Refinamiento de Modelos

### Dividir los datos en conjuntos de entrenamiento y prueba

1. Dividir los datos en conjuntos de entrenamiento y prueba.
2. Imprimir el número de muestras en los conjuntos de prueba y entrenamiento.


```python
X = df[features]
Y = df['price']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)

print("number of test samples:", x_test.shape[0])
print("number of training samples:", x_train.shape[0])
```

### Ajustar y evaluar un modelo de regresión Ridge

1. Crear y ajustar un modelo de regresión Ridge usando los datos de entrenamiento.
2. Calcular el R² utilizando los datos de prueba.


```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
print("R^2 de Ridge en datos de prueba:", ridge.score(x_test, y_test))
```

### Ajustar un modelo de regresión Ridge con características polinómicas

1. Realizar una transformación polinómica de segundo orden en los datos de entrenamiento y prueba.
2. Crear y ajustar un modelo de regresión Ridge con las características polinómicas.
3. Calcular el R² utilizando los datos de prueba.

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

ridge.fit(x_train_poly, y_train)
print("R^2 de Ridge con características polinómicas en datos de prueba:", ridge.score(x_test_poly, y_test))

```
